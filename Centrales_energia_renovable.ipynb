```python
def convert(df):
    if df["Eólica (MW)"] == 0.0:
        return False
    else: 
        return df["Eólica (MW)"]
```

```python
df["Eólica (MW) nueva"] = df.apply(convert, axis=1)
```

```python
data = df[df.País == "México"]
data = data[data["Eólica (MW) nueva"] != False]
```
